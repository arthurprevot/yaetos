```python
import panel as pn
import pydeck as pdk
pn.extension('deckgl')
```

```python
import pydeck as pdk

CLIMATE_TRACE_DATA = inputs['assets'][['longitude', 'latitude']]

# Define a layer to display on a map
layer = pdk.Layer(
    'HexagonLayer',
    CLIMATE_TRACE_DATA,
    # get_position=['lng', 'lat'],
    get_position=['longitude', 'latitude'],
    auto_highlight=True,
    elevation_scale=50,
    pickable=True,
    elevation_range=[0, 3000],
    extruded=True,                 
    coverage=1)

# Set the viewport location
view_state = pdk.ViewState(
    longitude=-100.415,
    latitude=32.2323,
    zoom=4,
    min_zoom=3,
    max_zoom=15,
    pitch=40.5,
    # bearing=-27.36)
    bearing=-0.36)

# Render
deck = pdk.Deck(layers=[layer], initial_view_state=view_state)
deck.to_html('./dashboards/demo_standalone.html')

# NOTE: doesn't require the mapbox API key. 
```

```python
# Layout
# row1_top = pn.Row(component_title)
# column1_row1 = pn.Row(component_deck)
# column1_row2 = pn.Row(component2)
# column1 = pn.Column(column1_row1, column1_row2)
# column2 = pn.Column(component3)
# row2_bottom = pn.Row(column1, column2)
# dashboard = pn.Row(row1_top, row2_bottom)

row2_column1 = pn.Column(row2_column1_content1, row2_column1_content2)
row2_column2 = pn.Column(row2_column2_content)
bottom_row = pn.Row(row2_column1, row2_column2)
dashboard = pn.Column(row1_content, bottom_row)
dashboard.servable()
```
