## Wikipedia Demo Dashboard Setup

This notebook defines the visualization to be served by 'Panel', the dashboarding library.
Integration with Yaetos is still experimental.

```python
from yaetos.etl_utils import InputLoader

args = {
    'job_param_file': 'conf/jobs_metadata.yml',
    'job_name': 'dashboards/wikipedia_demo_dashboard.ipynb',
    'py_job': 'n/a',
    'dependencies': False,
    }

inputs = InputLoader(**args).run()
```

```python
import panel as pn
pn.extension('tabulator')
```

```python
# Content
row1_content = pn.pane.Markdown("### Demo Dashboard\nWork In Progress !")
row2_column1_content1 = pn.widgets.Tabulator(inputs['ex1_sql'])
row2_column1_content2 = pn.pane.Markdown("A section here")
row2_column2_content = pn.pane.Markdown("A section here")
```

```python
# Layout
row2_column1 = pn.Column(row2_column1_content1, row2_column1_content2)
row2_column2 = pn.Column(row2_column2_content)
bottom_row = pn.Row(row2_column1, row2_column2)
dashboard = pn.Column(row1_content, bottom_row)
dashboard.servable()
```
